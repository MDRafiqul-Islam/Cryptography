```python
n_o_b = int(input("please enter number of bits: "))
```

```python
p=Crypto.Util.number.getPrime(n_o_b, randfunc=Crypto.Random.get_random_bytes)
q=Crypto.Util.number.getPrime(n_o_b, randfunc=Crypto.Random.get_random_bytes)
print('The value of first prime is:', p)
print('The value of seconf prime is:', q)

n = p * q
print('The value of n is:', n)

phi = (p-1) * (q-1)
if phi < 1 or phi > n:
  raise ValueError('phi_n cannot extend the limit')
print('The value of phi_n is:', phi)
```

```python
def gcd(a, b):
    while b:
        a, b = b, a%b
    return a
```

```python
def modInverse(x, p):
  y = pow(x, -1, p)
  return y
```

```python
e = random.randint(2,5)
while(True):
  co_prime=gcd(e,phi)
  if co_prime==1:
      break
  else:
      e = e + 1
d= modInverse (e,phi)

print('The value of Public key is:', e)
print('The value of Private key is:', d)
```

```python
# l=input("What is your message: ")
# res = ''.join(format(ord(i), '08b') for i in l)
# print(res)
# message = int(res, 2)
message=int(input("What is your message: "))
print('The message is: ',message)
```

```python
mix_server = int(input('Please Give the number of mix server'))
r = []
for i in range(mix_server):
  ran_value = random.randint(2, 2**n_o_b)
  if ran_value % 2 != 0:
    ran_value = ran_value + 1 
  r.append(ran_value)
  print('The value of random number', i, '=', r[i])
```

```python
pub = []
pri = []
pub.append(e)
pri.append(d)
for i in range(1,mix_server):
 temp = pub[i - 1] + 1
 while(True):
  co_prime=gcd(temp,phi)
  if co_prime==1:
      break
  else:
      temp=temp+1
 tempD=modInverse(temp,phi)
 pub.append(temp)
 pri.append(tempD)

for i in range(mix_server):
 print('The value of Public key of', i, 'is:', pub[i])
 print('The value of Private key of', i, 'is:', pri[i])
```

```python
message = message ^ r[0]
c = pow(message, e, n)
print('After 1st time encryption the value of encryption is: ',c)
# c =int(c ^ r[0]) 
print('After 1st time mixing the value of encryption is: ',c)

for i in range(1, mix_server):
 c = int(c ^ r[i])
 c = pow(c, pub[i], n)
 print('After', i+1,'time encryption the value of encryption is: ',c)
#  c = int(c ^ r[i])
 print('After', i+1,'time mixing the value of encryption is: ',c, '\n')

print('\n \n')
print('After', mix_server,'times the encryption value is', c)
```
