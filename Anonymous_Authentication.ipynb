```python
from random import randrange, random
import random
import hashlib
from random import randint
import numpy as np
```

```python
def encryption(a,b):
  d = a ^ b
  d = hash(d)
  return d
```

```python
ID = []
n_o_b = int(input('Please select number of bit: '))
n_o = int(input('Please input number of ID: '))
for i in range(n_o):
    ids = random.randint(2, 2**n_o_b)
    ID.append(ids) 
    print('ID of client', i, '=', ID[i])
j = random.randint(0, n_o - 1)
C_id = ID[j]
print('Client ID location = ', j)
```

```python
Pas = []
for i in range(len(ID)):
    ids = random.randint(2, 2**n_o_b)
    Pas.append(ids)
    print('Password of client', i, '=', Pas[i]) 
q = Pas[j]      
print('Client Password is = ', q)
```

```python
D = random.sample(ID, k=int(input("NO of id you want to sent: ")))
c = 0
for i in range(len(D)):
  if D[i] == C_id:
    c = 1
if c == 0:
  D.pop(randrange(len(D)))
  D.insert(random.randint(0, len(D)), C_id)
for i in range(len(D)):
    print('ID of client', i, '=', D[i])
```

```python
ind = []
for i in range(len(D)):
  aps = ID.index(D[i])
  ind.append(aps)
```

```python
P = []
for i in range (len(Pas)):
  for j in range (len(Pas)):
    if j == ind[i]:
      P.append(Pas[j])
  if len(ind) == len(P):
    break
for i in range(len(D)):
    print('Password of client', i, '=', P[i]) 
```

```python
for i in range(len(P)):
  P[i] = Q[i] ^ r
  print('Replace Password',i,'is:', P[i])
```
